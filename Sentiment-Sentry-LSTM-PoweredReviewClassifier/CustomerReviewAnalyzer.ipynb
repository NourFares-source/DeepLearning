```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

```python
# 1. DATA LOADING & SPLITTING (IMDb Dataset)
# ==========================================

MAX_VOCAB = 10000
MAX_LEN = 200 # We'll look at the first 200 words of each review
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.imdb.load_data(num_words=MAX_VOCAB)
```

```python
#Manual Padding
# This ensures every review is exactly 200 integers long.
x_train = tf.keras.preprocessing.sequence.pad_sequences(x_train, maxlen=MAX_LEN)
x_test = tf.keras.preprocessing.sequence.pad_sequences(x_test, maxlen=MAX_LEN)
```

```python
# 2. MODEL ARCHITECTURE

def build_lstm_sentry():
    model = models.Sequential([
        # Problem 4: Embedding
        layers.Embedding(input_dim=MAX_VOCAB, output_dim=128, input_length=MAX_LEN),

        # dropout/recurrent_dropout helps prevent overfitting
        layers.LSTM(64, dropout=0.2, recurrent_dropout=0.2),

        # Dense layers for final decision making
        layers.Dense(32, activation='relu'),
        layers.Dropout(0.5), # Another guard against overfitting
        layers.Dense(1, activation='sigmoid') # 0 = Negative, 1 = Positive
    ])
    model.compile(
        optimizer='adam',
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model
```
